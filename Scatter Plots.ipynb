```python
# Set the x-axis to a list of strings for each month.
y_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
x_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
If we were only to use the plt.plot() function with our x-axis and y-axis
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(y_axis, x_axis, c='red', label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend(loc=1)
plt.xlim([0,50])
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)

```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis_larger, color="blue")
```

```python
plt.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color="skyblue",alpha=0.8,edgecolor="black",linewidths=2, label = 'Boston')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend()
plt.xlim([0,50])
plt.gca().invert_yaxis()
```
